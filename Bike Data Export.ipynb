Combined Data for Membership in all Quarters

```python
q1_file = "TripData2018/Q1Bike2018.csv"
q2_file = "TripData2018/Q2Bike2018.csv"
q3_file = "TripData2018/Q3Bike2018.csv"
q4_file = "TripData2018/Q4Bike2018.csv"

q1 = pd.read_csv(q1_file)
q2 = pd.read_csv(q2_file)
q3 = pd.read_csv(q3_file)
q4 = pd.read_csv(q4_file)

q1.head()
```

Combine Files for Trip Data

```python
file_name1 = "TripData2018/JC-201801-citibike-tripdata.csv"
file_name2 = "TripData2018/JC-201802-citibike-tripdata.csv"
file_name3 = "TripData2018/JC-201803-citibike-tripdata.csv"
file_name4 = "TripData2018/JC-201804-citibike-tripdata.csv"
file_name5 = "TripData2018/JC-201805-citibike-tripdata.csv"
file_name6 = "TripData2018/JC-201806-citibike-tripdata.csv"
file_name7 = "TripData2018/JC-201807-citibike-tripdata.csv"
file_name8 = "TripData2018/JC-201808-citibike-tripdata.csv"
file_name9 = "TripData2018/JC-201809-citibike-tripdata.csv"
file_name10 = "TripData2018/JC-201810-citibike-tripdata.csv"
file_name11 = "TripData2018/JC-201811-citibike-tripdata.csv"
file_name12 = "TripData2018/JC-201812-citibike-tripdata.csv"

january_trip_data = pd.read_csv(file_name1)
february_trip_data = pd.read_csv(file_name2)
march_trip_data = pd.read_csv(file_name3)
april_trip_data = pd.read_csv(file_name4)
may_trip_data = pd.read_csv(file_name5)
june_trip_data = pd.read_csv(file_name6)
july_trip_data = pd.read_csv(file_name7)
august_trip_data = pd.read_csv(file_name8)
september_trip_data = pd.read_csv(file_name9)
october_trip_data = pd.read_csv(file_name10)
november_trip_data = pd.read_csv(file_name11)
december_trip_data = pd.read_csv(file_name12)

january_trip_data.head()
```

```python
month_frames = [january_trip_data, february_trip_data, march_trip_data, april_trip_data,
           may_trip_data, june_trip_data, july_trip_data, august_trip_data,
           september_trip_data, october_trip_data, november_trip_data, december_trip_data]
```

```python
month_data = pd.concat(month_frames)
```

```python
month_data.to_csv("Trip_Data_2018.csv")
```
